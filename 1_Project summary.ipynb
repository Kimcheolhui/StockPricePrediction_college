# 프로젝트 주제 선정 이유

이번 코로나 사태로 인해 많은 기업들의 주가가 하락했습니다.   

제가 가지고 있는 기업들의 주가도 떨어지면서 많은 손해을 봤었던 기억이 있습니다.   

그런 기억을 가지고 소활코 수업 내용 중에 딥러닝 분야에 대해 배우면서, '주가를 예측한다면?'이라는 생각을 해보았습니다.   

물론 주가를 예측한다는 것이 굉장히 어려운 일이고, 저 뿐만 아니라 많은 사람들이 시도해봤을 거라고 생각했습니다.   

하지만 모든 자료들이 주가 자체를 예측하는 것이었고, 주가를 예측하는 게 아닌 주가가 오르는지 안 오르는지만 예측하는 자료는 단 하나도 없었습니다.   

따라서, 인터넷 상에 올라와있는 것들과 겹치지 않으면서도 제가 평소 관심있게 생각했던 분야로 기말 프로젝트를 진행할 수 있을 것 같다는 생각에 이를 주제로 선정하게 되었습니다.

# 과제 수행 후 알게 된 점, 느낀 점

첫 번째, 코딩 능력이 매우 중요하다는 것입니다. 어느정도 공부를 했어도 제가 모르는 게 많다는 것을 느꼈습니다. 이 프로젝트를 진행하면서, 코드 한 줄을 짜는 데에도 확신이 서지 않아서 한 번 돌려보고, cell을 하나하나 구분하고, 인터넷에 검색했던 적이 많았습니다. 어떠한 오류가 나면, 왜 났는지 몰라서 인터넷에 검색도 해보고, 오류가 났을 법한 부분을 하나하나 읽어보고, 그것도 안 되면 코딩을 잘하는 친구에게 커피를 사주고 질문해가면서 오류를 수정했습니다. 그런 과정을 겪으면서 저 스스로도 많은 성장을 했지만, 굉장히 많은 시간을 소모했던 것 같습니다. 사실, 완성이라고 말하기엔 부족한 이 프로젝트를 완성하는 데 약 2주라는 긴 시간을 썼던 것 같습니다. 주말을 모두 투자하기도 하고, 도서관 근로 중간 중간에 코드 몇 줄 적기도 하고, 아침 일찍 일어나서 밤까지 기숙사 책상에 앉아 주피터 노트북만 들어다보고 있었던 적도 있습니다. 많이 힘들기도 했지만, 그래도 나름 값진 경험이었던 것 같습니다.

두 번째, 데이터의 중요성입니다. 제가 딥러닝을 진행하면서 가장 부족하다고 느꼈던 점이 데이터였습니다. 삼성전자의 액면분할 때문에, 제한된 기간 동안의 데이터만 존재했습니다. 그에 따라 주가 데이터가 제가 생각했던 것보다 부족했던 것 같습니다. 또한, 단순히 수집한 데이터 그 자체로는 딥러닝을 할 수 없다는 것이었습니다. 형식이 맞지 않아서 문자 class를 숫자 class로 바꾸는 것이라든지, feature를 수정한다든지, 중간중간 빈 데이터를 지워준다든지, 여러가지 전처리 과정을 거치면서 역시 데이터의 중요성을 느꼈습니다.

세 번째, 딥러닝에 여러 방법이 있고, 저는 아직 아무것도 모른다는 것을 알게 됐습니다. 수업 시간에 배운 딥러닝 내용만 가지고 주가의 상승 하락 여부를 예측하기에는 부족하다는 것을 깨달았고, 그 과정 속에서 시계열 데이터라는 것에 대해 알게 되었습니다. 또한 RNN과 LSTM에 대해서도 알게 되었습니다. 그것을 배워가는 과정에서 수많은 용어들이 나오는 것을 볼 수 있었고, '내가 아는 건 정말 세 발의 피구나'하는 생각이 들었습니다. 

네 번째, '컴퓨터를 진로로 선택하길 잘했다.'입니다. 중간 프로젝트에서도 제가 말씀 드렸듯이, 저는 현재 생명과학전공입니다. 하지만 적성에 맞지 않아 현재 전전컴 전공으로 전과를 신청해놓은 상태이고, 다음 학기부터 전전컴 전공을 수업을 듣게 됩니다. 이번 기말 프로젝트를 진행하면서, '내 손으로 만드는' 것에 다시 한 번 더 흥미를 느끼게 되었고, 중간 프로젝트 때와 마찬가지로 컴퓨터 분야에 대한 진로를 확신할 수 있었습니다.

별개로, 제가 이번 겨울에 기회가 된다면 '지스트 졸업 요건' 프로그램을 만들어볼 생각인데, 시간 여유가 된다면 학점 예측 모델도 한 번 만들어볼까 생각 중입니다. 기대가 됩니다.

# 아쉬웠던 점

많은 시간을 투자했고 열정을 쏟았지만, 뿌듯함에 앞서 딥러닝 결과가 그닥 좋지 않다는 사실이 가장 아쉬운 것 같습니다. 기말 시험이 끝나면 보완을 해봐야 할 것 같습니다.

지금 이 글을 쓰면서 생각난 게 있는데, 그냥 주가 예측으로 딥러닝을 진행하고, 그 전날의 예측 값과 비교해서 더 크면 올랐다, 작으면 떨어졌다라고 했으면 좀 더 좋은 결과가 있지 않았을까 싶습니다.
